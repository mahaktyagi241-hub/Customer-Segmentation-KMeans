# Customer Segmentation using K-Means Clustering

## Objective

The objective of this project is to segment mall customers into different groups based on their Annual Income and Spending Score using the K-Means Clustering algorithm.

## Import Libraries

## Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("Mall_Customers.csv")

df.head()

```

## Data Exploration

```python
# Number of rows and columns
print("Shape:", df.shape)

# Information about dataset
df.info()
```

### Statistical Summary

```python
df.describe()
```

### Missing Values Check

```python
df.isnull().sum()
```

### Age Distribution

```python
plt.figure(figsize=(6,4))
plt.hist(df["Age"], bins=10)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")

plt.show()


```

### Annual Income vs Spending Score

```python
plt.figure(figsize=(7,5))

plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"])

plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")

plt.show()
```

## Choosing the Optimal Number of Clusters

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(df[["Annual Income (k$)", "Spending Score (1-100)"]])
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(7,5))
plt.plot(range(1, 11), wcss, marker='o')

plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.show()
```

From the elbow graph, the bend occurs around **5 clusters**, so K = 5 is selected for the K-Means algorithm.

## Applying K-Means Clustering

```python
# Sirf Income aur Spending Score ko use karenge
X = df[["Annual Income (k$)", "Spending Score (1-100)"]]

# K-Means Model
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)

# Model Train + Prediction
df["Cluster"] = kmeans.fit_predict(X)

# Pehli 5 rows dekho
df.head()
```

## Final Customer Segmentation

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X["Annual Income (k$)"],
    X["Spending Score (1-100)"],
    c=df["Cluster"],
    cmap="viridis",
    s=60
)

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    c="red",
    marker="X",
    s=200,
    label="Centroids"
)

plt.title("Customer Segmentation using K-Means")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()

plt.show()
```

```python
df.groupby("Cluster")[["Annual Income (k$)", "Spending Score (1-100)", "Age"]].mean()
```
